```python
import pandas as pd
from googleapiclient.discovery import build

# 🔐 Replace with your API key
API_KEY = "AIzaSyClEBK38SAUlWFfTWjgR5WG-tSKTdsWAYs"
VIDEO_ID = "dQw4w9WgXcQ"

# Build YouTube API client
youtube = build("youtube", "v3", developerKey=API_KEY)

authors = []
likes = []

def fetch_comments(video_id, max_comments=300):
    next_page_token = None

    while len(authors) < max_comments:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,
            pageToken=next_page_token
        )

        response = request.execute()

        for item in response.get('items', []):
            snippet = item['snippet']['topLevelComment']['snippet']

            author = snippet.get('authorDisplayName', "Unknown")
            like = snippet.get('likeCount', 0)

            authors.append(author)
            likes.append(like)

        next_page_token = response.get('nextPageToken')

        if not next_page_token:
            break

# Fetch comments
fetch_comments(VIDEO_ID)

# Create DataFrame
df = pd.DataFrame({
    "Author": authors,
    "Likes": likes
})

# Save CSV
df.to_csv("youtube_comments_dashboard.csv", index=False)

print("✅ Data saved successfully!")
print(df.head())
```
