```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
df1 = pd.read_csv("apple_dataset.csv")   # Apple
df2 = pd.read_csv("samsung_dataset.csv") # Samsung

# Add labels
df1['competitor'] = 'Apple'
df2['competitor'] = 'Samsung'

# Combine datasets
df = pd.concat([df1, df2], ignore_index=True)

# Select relevant columns
analysis_df = df[['competitor', 'followersCount', 'postsCount', 'followsCount']]

# Group by competitor and take mean
comparison = analysis_df.groupby('competitor').mean()

# -------------------------------
# VISUALIZATIONS
# -------------------------------

# Followers Comparison (Bar)
plt.figure()
comparison['followersCount'].plot(kind='bar')
plt.title("Followers Comparison (Apple vs Samsung)")
plt.ylabel("Followers")
plt.show()

# Posts Comparison (Bar)
plt.figure()
comparison['postsCount'].plot(kind='bar')
plt.title("Posts Count Comparison")
plt.ylabel("Posts")
plt.show()

# Market Share (Pie)
plt.figure()
comparison['followersCount'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Market Share (Followers)")
plt.ylabel("")
plt.show()

# Line Plot (Overall Trend)
plt.figure()
comparison.plot(kind='line', marker='o')
plt.title("Overall Comparison Trend")
plt.ylabel("Values")
plt.show()

# Stacked Bar Chart
plt.figure()
comparison.plot(kind='bar', stacked=True)
plt.title("Combined Metrics Comparison")
plt.ylabel("Values")
plt.show()

# Horizontal Bar Chart
plt.figure()
comparison['followersCount'].plot(kind='barh')
plt.title("Followers Comparison (Horizontal)")
plt.xlabel("Followers")
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(analysis_df['postsCount'], analysis_df['followersCount'])
plt.xlabel("Posts Count")
plt.ylabel("Followers Count")
plt.title("Posts vs Followers Relationship")
plt.show()
```
